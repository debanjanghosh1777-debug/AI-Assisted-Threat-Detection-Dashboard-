#Create a simple Python project in Google Colab using list, tuple, dictionary, OOPs, NumPy, Pandas and Matplotlib. The project should be based on a banking transaction dataset where the system checks transaction amount, failed login attempts, unknown location and new device to detect possible threats. Create a risk score, threat status and show simple visualizations using bar graph, pie chart and line graph.

```python
# AI Assisted Banking Threat Detection Dashboard

# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1. List
# -------------------------------

customers = ["Amit", "Riya", "John", "Priya", "Rahul"]

print("Customer List:")
print(customers)


# -------------------------------
# 2. Tuple
# -------------------------------

project_info = ("AI Assisted Banking Threat Detection Dashboard", "Python Project")

print("\nProject Information:")
print("Project Name:", project_info[0])
print("Project Type:", project_info[1])


# -------------------------------
# 3. Dictionary
# -------------------------------

risk_rules = {
    "High_Amount_Limit": 50000,
    "Failed_Login_Limit": 3,
    "Risk_Location": "Unknown",
    "Risk_Device": "New Device"
}

print("\nRisk Rules:")
print(risk_rules)


# -------------------------------
# 4. Pandas DataFrame
# -------------------------------

data = {
    "Customer": ["Amit", "Riya", "John", "Priya", "Rahul"],
    "Transaction_Amount": [12000, 75000, 30000, 90000, 15000],
    "Failed_Login_Attempts": [1, 4, 2, 5, 0],
    "Location": ["India", "Unknown", "India", "Unknown", "India"],
    "Device": ["Mobile", "New Device", "Laptop", "New Device", "Mobile"],
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
}

df = pd.DataFrame(data)

print("\nBanking Transaction Data:")
display(df)


# -------------------------------
# 5. OOPs Class for Threat Detection
# -------------------------------

class ThreatDetector:

    def calculate_risk_score(self, row):
        score = 0

        if row["Transaction_Amount"] > risk_rules["High_Amount_Limit"]:
            score = score + 40

        if row["Failed_Login_Attempts"] >= risk_rules["Failed_Login_Limit"]:
            score = score + 30

        if row["Location"] == risk_rules["Risk_Location"]:
            score = score + 20

        if row["Device"] == risk_rules["Risk_Device"]:
            score = score + 10

        return score

    def threat_status(self, score):
        if score >= 70:
            return "High Risk"
        elif score >= 40:
            return "Medium Risk"
        else:
            return "Low Risk"


# Creating object of class
detector = ThreatDetector()

# Applying risk score function
df["Risk_Score"] = df.apply(detector.calculate_risk_score, axis=1)

# Applying threat status function
df["Threat_Status"] = df["Risk_Score"].apply(detector.threat_status)

print("\nFinal Threat Detection Data:")
display(df)


# -------------------------------
# 6. NumPy Calculation
# -------------------------------

amounts = np.array(df["Transaction_Amount"])

print("\nNumPy Calculations:")
print("Average Transaction Amount:", np.mean(amounts))
print("Highest Transaction Amount:", np.max(amounts))
print("Lowest Transaction Amount:", np.min(amounts))


# -------------------------------
# 7. Data Visualization
# -------------------------------

# Bar Graph - Customer-wise Risk Score
plt.figure(figsize=(8,5))
plt.bar(df["Customer"], df["Risk_Score"])
plt.title("Customer-wise Risk Score")
plt.xlabel("Customer")
plt.ylabel("Risk Score")
plt.show()


# Pie Chart - Threat Status Count
threat_count = df["Threat_Status"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(threat_count, labels=threat_count.index, autopct="%1.1f%%")
plt.title("Threat Status Distribution")
plt.show()


# Line Graph - Transaction Amount by Day
plt.figure(figsize=(8,5))
plt.plot(df["Day"], df["Transaction_Amount"], marker="o")
plt.title("Transaction Amount by Day")
plt.xlabel("Day")
plt.ylabel("Transaction Amount")
plt.show()
```

#Create a professional Python project in Google Colab for an AI-style threat detection system. Use list, tuple, dictionary, OOPs, NumPy, Pandas and Matplotlib. Create a sample user activity dataset. Identify risks based on failed login attempts, high transaction amount, high data transfer, unknown location, new device usage and suspicious activity. Calculate risk score, classify risk level and create professional visualizations such as bar chart, pie chart, line chart, scatter plot and heatmap.

```python
# AI Threat Risk Detection and Visualization System

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------
# 1. List
# ---------------------------------------

threat_activities = [
    "Malware Alert",
    "Network Scan",
    "Suspicious Transfer",
    "Data Exfiltration"
]

print("Threat Activities List:")
print(threat_activities)


# ---------------------------------------
# 2. Tuple
# ---------------------------------------

project_info = (
    "AI Threat Risk Detection System",
    "Python Data Analytics Project",
    "Google Colab"
)

print("\nProject Information:")
print("Project Name:", project_info[0])
print("Project Type:", project_info[1])
print("Platform:", project_info[2])


# ---------------------------------------
# 3. Dictionary
# ---------------------------------------

risk_weights = {
    "Failed_Login": 25,
    "High_Transaction": 20,
    "High_Data_Transfer": 20,
    "Unknown_Location": 15,
    "New_Device": 10,
    "Suspicious_Activity": 25,
    "Odd_Login_Hour": 10
}

print("\nRisk Weight Rules:")
print(risk_weights)


# ---------------------------------------
# 4. Sample Data using Pandas
# ---------------------------------------

data = {
    "User": ["Amit", "Riya", "John", "Priya", "Rahul", "Neha", "David", "Sara", "Manish", "Kiran"],
    "Activity": ["Login Success", "Failed Login", "Network Scan", "Suspicious Transfer", "File Access",
                 "Malware Alert", "Login Success", "Failed Login", "Data Download", "Suspicious Transfer"],
    "Failed_Logins": [0, 6, 2, 4, 1, 5, 0, 7, 1, 3],
    "Transaction_Amount": [12000, 75000, 5000, 90000, 8000, 30000, 15000, 60000, 10000, 75000],
    "Data_Transfer_MB": [80, 120, 700, 650, 60, 800, 90, 300, 900, 550],
    "Location": ["India", "India", "India", "Unknown", "India", "Unknown", "India", "Unknown", "India", "Unknown"],
    "Device": ["Mobile", "New Device", "Laptop", "New Device", "Desktop", "New Device", "Mobile", "Laptop", "New Device", "New Device"],
    "Login_Hour": [10, 23, 1, 2, 14, 3, 11, 22, 16, 4],
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday"]
}

df = pd.DataFrame(data)

print("\nOriginal User Activity Data:")
display(df)


# ---------------------------------------
# 5. NumPy for Automatic Thresholds
# ---------------------------------------

transaction_threshold = np.percentile(df["Transaction_Amount"], 75)
data_transfer_threshold = np.percentile(df["Data_Transfer_MB"], 75)

print("\nAutomatic Thresholds:")
print("High Transaction Threshold:", transaction_threshold)
print("High Data Transfer Threshold:", data_transfer_threshold)


# ---------------------------------------
# 6. OOPs Threat Detection Class
# ---------------------------------------

class ThreatAnalyzer:

    def calculate_risk_score(self, row):
        score = 0

        if row["Failed_Logins"] >= 5:
            score = score + risk_weights["Failed_Login"]

        if row["Transaction_Amount"] > transaction_threshold:
            score = score + risk_weights["High_Transaction"]

        if row["Data_Transfer_MB"] > data_transfer_threshold:
            score = score + risk_weights["High_Data_Transfer"]

        if row["Location"] == "Unknown":
            score = score + risk_weights["Unknown_Location"]

        if row["Device"] == "New Device":
            score = score + risk_weights["New_Device"]

        if row["Activity"] in threat_activities:
            score = score + risk_weights["Suspicious_Activity"]

        if row["Login_Hour"] >= 22 or row["Login_Hour"] <= 5:
            score = score + risk_weights["Odd_Login_Hour"]

        if score > 100:
            score = 100

        return score


    def classify_risk(self, score):
        if score >= 75:
            return "Critical"
        elif score >= 50:
            return "High"
        elif score >= 30:
            return "Medium"
        else:
            return "Low"


    def recommend_action(self, risk_level):
        if risk_level == "Critical":
            return "Block and Investigate"
        elif risk_level == "High":
            return "Investigate Immediately"
        elif risk_level == "Medium":
            return "Monitor"
        else:
            return "Allow"


# Creating object
analyzer = ThreatAnalyzer()

# Applying threat analysis
df["Risk_Score"] = df.apply(analyzer.calculate_risk_score, axis=1)
df["Risk_Level"] = df["Risk_Score"].apply(analyzer.classify_risk)
df["Recommended_Action"] = df["Risk_Level"].apply(analyzer.recommend_action)

print("\nFinal Threat Analysis Data:")
display(df)


# ---------------------------------------
# 7. Summary Report
# ---------------------------------------

print("\nThreat Summary Report:")
print("Total Records:", len(df))
print("Average Risk Score:", np.mean(df["Risk_Score"]))
print("Highest Risk Score:", np.max(df["Risk_Score"]))
print("Lowest Risk Score:", np.min(df["Risk_Score"]))
print("Critical Threats:", len(df[df["Risk_Level"] == "Critical"]))
print("High Risk Threats:", len(df[df["Risk_Level"] == "High"]))


# ---------------------------------------
# 8. Data Visualizations
# ---------------------------------------

# Chart 1: User-wise Risk Score Bar Chart

plt.figure(figsize=(10,6))
plt.bar(df["User"], df["Risk_Score"], color="crimson")
plt.title("User-wise Risk Score")
plt.xlabel("Users")
plt.ylabel("Risk Score")
plt.xticks(rotation=45)
plt.show()


# Chart 2: Risk Level Distribution Pie Chart

risk_count = df["Risk_Level"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(risk_count, labels=risk_count.index, autopct="%1.1f%%", startangle=90)
plt.title("Risk Level Distribution")
plt.show()


# Chart 3: Day-wise Average Risk Line Chart

day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

day_risk = df.groupby("Day")["Risk_Score"].mean()
day_risk = day_risk.reindex(day_order)

plt.figure(figsize=(10,6))
plt.plot(day_risk.index, day_risk.values, marker="o", linewidth=2)
plt.title("Day-wise Average Risk Trend")
plt.xlabel("Day")
plt.ylabel("Average Risk Score")
plt.xticks(rotation=45)
plt.show()


# Chart 4: Failed Logins vs Data Transfer Scatter Plot

plt.figure(figsize=(9,6))
plt.scatter(
    df["Failed_Logins"],
    df["Data_Transfer_MB"],
    s=df["Risk_Score"] * 5 + 50,
    c=df["Risk_Score"],
    cmap="Reds",
    edgecolors="black"
)

plt.colorbar(label="Risk Score")
plt.title("Failed Logins vs Data Transfer Risk Analysis")
plt.xlabel("Failed Login Attempts")
plt.ylabel("Data Transfer MB")
plt.show()


# Chart 5: Heatmap for Day-wise Risk Level Count

risk_order = ["Low", "Medium", "High", "Critical"]

heatmap_data = pd.crosstab(df["Day"], df["Risk_Level"])
heatmap_data = heatmap_data.reindex(index=day_order, columns=risk_order, fill_value=0)

plt.figure(figsize=(8,6))
plt.imshow(heatmap_data, cmap="Reds")

plt.title("Threat Heatmap: Day-wise Risk Level Count")
plt.xlabel("Risk Level")
plt.ylabel("Day")

plt.xticks(range(len(risk_order)), risk_order)
plt.yticks(range(len(day_order)), day_order)

for i in range(len(day_order)):
    for j in range(len(risk_order)):
        plt.text(j, i, heatmap_data.iloc[i, j], ha="center", va="center")

plt.colorbar(label="Threat Count")
plt.show()
```
