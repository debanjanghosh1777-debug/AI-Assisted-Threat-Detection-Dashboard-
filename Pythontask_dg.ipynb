#Create a simple Python program using OOPs. Create a Student class with name, age and course. Add a function to display student details.

```python
# OOPs example in Python

class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def display_details(self):
        print("Student Details:")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Course:", self.course)

student1 = Student("Rahul", 22, "MBA")

student1.display_details()
```

#Create a simple Python project in Google Colab using list, tuple, dictionary, OOPs, NumPy, Pandas and Matplotlib. The project should be based on a banking transaction dataset where the system checks transaction amount, failed login attempts, unknown location and new device to detect possible threats. Create a risk score, threat status and show simple visualizations using bar graph, pie chart and line graph.

```python
# AI Assisted Banking Threat Detection Dashboard

# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1. List
# -------------------------------

customers = ["Amit", "Riya", "John", "Priya", "Rahul"]

print("Customer List:")
print(customers)


# -------------------------------
# 2. Tuple
# -------------------------------

project_info = ("AI Assisted Banking Threat Detection Dashboard", "Python Project")

print("\nProject Information:")
print("Project Name:", project_info[0])
print("Project Type:", project_info[1])


# -------------------------------
# 3. Dictionary
# -------------------------------

risk_rules = {
    "High_Amount_Limit": 50000,
    "Failed_Login_Limit": 3,
    "Risk_Location": "Unknown",
    "Risk_Device": "New Device"
}

print("\nRisk Rules:")
print(risk_rules)


# -------------------------------
# 4. Pandas DataFrame
# -------------------------------

data = {
    "Customer": ["Amit", "Riya", "John", "Priya", "Rahul"],
    "Transaction_Amount": [12000, 75000, 30000, 90000, 15000],
    "Failed_Login_Attempts": [1, 4, 2, 5, 0],
    "Location": ["India", "Unknown", "India", "Unknown", "India"],
    "Device": ["Mobile", "New Device", "Laptop", "New Device", "Mobile"],
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
}

df = pd.DataFrame(data)

print("\nBanking Transaction Data:")
display(df)


# -------------------------------
# 5. OOPs Class for Threat Detection
# -------------------------------

class ThreatDetector:

    def calculate_risk_score(self, row):
        score = 0

        if row["Transaction_Amount"] > risk_rules["High_Amount_Limit"]:
            score = score + 40

        if row["Failed_Login_Attempts"] >= risk_rules["Failed_Login_Limit"]:
            score = score + 30

        if row["Location"] == risk_rules["Risk_Location"]:
            score = score + 20

        if row["Device"] == risk_rules["Risk_Device"]:
            score = score + 10

        return score

    def threat_status(self, score):
        if score >= 70:
            return "High Risk"
        elif score >= 40:
            return "Medium Risk"
        else:
            return "Low Risk"


# Creating object of class
detector = ThreatDetector()

# Applying risk score function
df["Risk_Score"] = df.apply(detector.calculate_risk_score, axis=1)

# Applying threat status function
df["Threat_Status"] = df["Risk_Score"].apply(detector.threat_status)

print("\nFinal Threat Detection Data:")
display(df)


# -------------------------------
# 6. NumPy Calculation
# -------------------------------

amounts = np.array(df["Transaction_Amount"])

print("\nNumPy Calculations:")
print("Average Transaction Amount:", np.mean(amounts))
print("Highest Transaction Amount:", np.max(amounts))
print("Lowest Transaction Amount:", np.min(amounts))


# -------------------------------
# 7. Data Visualization
# -------------------------------

# Bar Graph - Customer-wise Risk Score
plt.figure(figsize=(8,5))
plt.bar(df["Customer"], df["Risk_Score"])
plt.title("Customer-wise Risk Score")
plt.xlabel("Customer")
plt.ylabel("Risk Score")
plt.show()


# Pie Chart - Threat Status Count
threat_count = df["Threat_Status"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(threat_count, labels=threat_count.index, autopct="%1.1f%%")
plt.title("Threat Status Distribution")
plt.show()


# Line Graph - Transaction Amount by Day
plt.figure(figsize=(8,5))
plt.plot(df["Day"], df["Transaction_Amount"], marker="o")
plt.title("Transaction Amount by Day")
plt.xlabel("Day")
plt.ylabel("Transaction Amount")
plt.show()
```
