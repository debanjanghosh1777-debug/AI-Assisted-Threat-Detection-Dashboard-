## INSTALL SQL


```python
import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect("security.db")

# Create fresh table
conn.execute("DROP TABLE IF EXISTS threats")

conn.execute("""
CREATE TABLE threats (
    user TEXT,
    activity TEXT,
    status TEXT,
    threat_level TEXT
)
""")

# Add data
conn.execute("""
INSERT INTO threats VALUES
('Rahul', 'Login', 'Success', 'Low'),
('Amit', 'Failed Login', 'Failed', 'High'),
('Unknown', 'SQL Injection', 'Blocked', 'Critical'),
('Sneha', 'File Access', 'Success', 'Low')
""")

# Show full table
print("All Data")
display(pd.read_sql_query("SELECT * FROM threats", conn))

# Show risky data
print("Risky Activities")
display(pd.read_sql_query("""
SELECT * FROM threats
WHERE threat_level = 'High' OR threat_level = 'Critical'
""", conn))

# Count threat levels
print("Threat Count")
display(pd.read_sql_query("""
SELECT threat_level, COUNT(*) AS total
FROM threats
GROUP BY threat_level
""", conn))
```
