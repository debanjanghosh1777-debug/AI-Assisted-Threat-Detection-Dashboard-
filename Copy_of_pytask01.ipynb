```python
# AI Assisted Threat Detection Dashboard - Simple Version

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create sample security activity data
data = {
    "User": ["Admin", "Rahul", "Priya", "Guest", "John", "System"],
    "Activity": ["Login Success", "Login Failed", "File Access", "Login Failed", "Network Scan", "Malware Alert"],
    "Source_IP": ["192.168.1.10", "192.168.1.20", "192.168.1.30", "45.23.12.88", "10.0.0.8", "103.44.21.90"],
    "Failed_Attempts": [0, 3, 0, 6, 0, 0]
}

# Step 2: Convert data into table
df = pd.DataFrame(data)

print("Security Log Data")
display(df)

# Step 3: Create risk score using normal function
def risk_score(row):
    if row["Activity"] == "Malware Alert":
        return 90
    elif row["Failed_Attempts"] >= 5:
        return 80
    elif row["Activity"] == "Network Scan":
        return 60
    elif row["Activity"] == "Login Failed":
        return 50
    else:
        return 20

df["Risk_Score"] = df.apply(risk_score, axis=1)

# Step 4: Create threat status using normal function
def threat_status(score):
    if score >= 70:
        return "High Risk"
    else:
        return "Normal"

df["Threat_Status"] = df["Risk_Score"].apply(threat_status)

print("Final Threat Detection Table")
display(df)

# Step 5: Create graph
plt.figure(figsize=(8,5))
plt.bar(df["User"], df["Risk_Score"])

plt.title("User-wise Risk Score")
plt.xlabel("User")
plt.ylabel("Risk Score")

plt.show()
```
